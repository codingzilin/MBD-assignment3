```python
import pandas as pd

test_data = pd.read_csv('/Users/miao/Documents/Mining Big Data/final assignment/Groceries data test.csv')
train_data = pd.read_csv('/Users/miao/Documents/Mining Big Data/final assignment/Groceries data train.csv')

test_data.head(), train_data.head()
```

```python
# Parsing Date Columns
train_data['Date'] = pd.to_datetime(train_data['Date'], format='%d/%m/%Y')

# Removing Duplicate Data
train_data = train_data.drop_duplicates()

# View processed data
train_data.head()
```

```python
# Building a user-commodity interaction matrix
user_item_matrix = train_data.pivot_table(index='Member_number', columns='itemDescription', aggfunc='size', fill_value=0)

# View the first few rows of the interaction matrix
user_item_matrix.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate the cosine similarity between goods
item_similarity = cosine_similarity(user_item_matrix.T)
item_similarity_df = pd.DataFrame(item_similarity, index=user_item_matrix.columns, columns=user_item_matrix.columns)

# View the first few rows of the similarity matrix
item_similarity_df.head()
```

```python
def get_top_k_similar_items(item_similarity_df, item, k=5):
    """
    Get the top k items that are most similar to the specified item.
    :param item_similarity_df: product similarity matrix
    :param item: the specified item
    :param k: top k similar items
    :return: list of similar products
    """
    if item not in item_similarity_df.index:
        return []
    
    # Get Similarity Sort
    similar_items = item_similarity_df[item].sort_values(ascending=False)
    
    # Exclude the product itself and return the first k similar products
    top_k_items = similar_items.iloc[1:k+1].index.tolist()
    return top_k_items

# Example: Get the top 5 products most similar to “whole milk”.
similar_items = get_top_k_similar_items(item_similarity_df, "whole milk", k=5)
similar_items
```

```python
# Reading test set data
test_data_path = '/Users/miao/Documents/Mining Big Data/final assignment/Groceries data test.csv'
test_data = pd.read_csv(test_data_path)

# Parsing Date Columns
test_data['Date'] = pd.to_datetime(test_data['Date'], format='%d/%m/%Y')

# Removing Duplicate Data
test_data = test_data.drop_duplicates()

# Example: Generate a list of recommendations for each user in the test set
test_user_item_matrix = test_data.pivot_table(index='Member_number', columns='itemDescription', aggfunc='size', fill_value=0)

# View the first few rows of the test set data
test_user_item_matrix.head()
```

```python
def generate_recommendations(user_item_matrix, item_similarity_df, k=5):
    recommendations = {}
    for user in user_item_matrix.index:
        user_items = user_item_matrix.columns[user_item_matrix.loc[user] > 0]
        user_recommendations = set()
        for item in user_items:
            similar_items = get_top_k_similar_items(item_similarity_df, item, k)
            user_recommendations.update(similar_items)
        recommendations[user] = list(user_recommendations)
    return recommendations

# Generate Recommended Lists
test_recommendations = generate_recommendations(test_user_item_matrix, item_similarity_df, k=5)

# Example: View a list of recommendations for a user
test_recommendations[1000]
```

```python
def calculate_accuracy(user_item_matrix, recommendations):
    total_relevant_items = 0
    total_recommended_items = 0
    total_correct_recommendations = 0
    
    for user in user_item_matrix.index:
        actual_items = set(user_item_matrix.columns[user_item_matrix.loc[user] > 0])
        recommended_items = set(recommendations[user])
        
        total_relevant_items += len(actual_items)
        total_recommended_items += len(recommended_items)
        total_correct_recommendations += len(actual_items & recommended_items)
    
    precision = total_correct_recommendations / total_recommended_items
    recall = total_correct_recommendations / total_relevant_items
    return precision, recall

# Calculate the accuracy of the recommended results
precision, recall = calculate_accuracy(test_user_item_matrix, test_recommendations)
precision, recall
```

It shows our algorithm works to some extent. But we can also try to improve the accuracy of the algorithm. Next we try to use user-based collaborative filtering to improve the accuracy of the algorithm.

```python
# Calculate the cosine similarity between users
user_similarity = cosine_similarity(user_item_matrix)
user_similarity_df = pd.DataFrame(user_similarity, index=user_item_matrix.index, columns=user_item_matrix.index)

# View the first few rows of the user similarity matrix
user_similarity_df.head()
```

```python
def generate_user_based_recommendations(user_item_matrix, user_similarity_df, test_users, k=5):
    recommendations = {}
    for user in test_users:
        if user not in user_similarity_df.index:
            continue
        similar_users = user_similarity_df[user].sort_values(ascending=False).index[1:k+1]
        user_recommendations = set()
        for similar_user in similar_users:
            similar_user_items = user_item_matrix.columns[user_item_matrix.loc[similar_user] > 0]
            user_recommendations.update(similar_user_items)
        user_recommendations = user_recommendations - set(user_item_matrix.columns[user_item_matrix.loc[user] > 0])
        recommendations[user] = list(user_recommendations)
    return recommendations

# Get the list of users in the test set
test_users = test_user_item_matrix.index

# Generate Recommended Lists
user_based_recommendations = generate_user_based_recommendations(user_item_matrix, user_similarity_df, test_users, k=5)

# Example: View a list of recommendations for a user
user_based_recommendations[1000]
```

```python
def generate_user_based_recommendations(user_item_matrix, user_similarity_df, test_users, k=5):
    recommendations = {}
    for user in test_users:
        recommendations[user] = []  # 初始化空推荐列表
        if user not in user_similarity_df.index:
            continue
        similar_users = user_similarity_df[user].sort_values(ascending=False).index[1:k+1]
        user_recommendations = set()
        for similar_user in similar_users:
            similar_user_items = user_item_matrix.columns[user_item_matrix.loc[similar_user] > 0]
            user_recommendations.update(similar_user_items)
        user_recommendations = user_recommendations - set(user_item_matrix.columns[user_item_matrix.loc[user] > 0])
        recommendations[user] = list(user_recommendations)
    return recommendations

# Generate Recommended Lists
user_based_recommendations = generate_user_based_recommendations(user_item_matrix, user_similarity_df, test_users, k=5)

# Calculate the accuracy of the recommended results
user_based_precision, user_based_recall = calculate_accuracy(test_user_item_matrix, user_based_recommendations)
user_based_precision, user_based_recall
```

```python
def combine_recommendations(item_based_recommendations, user_based_recommendations):
    combined_recommendations = {}
    for user in item_based_recommendations:
        item_based_set = set(item_based_recommendations[user])
        user_based_set = set(user_based_recommendations[user])
        combined_set = item_based_set | user_based_set  # 取并集
        combined_recommendations[user] = list(combined_set)
    return combined_recommendations

# Generate item-based recommendation lists
item_based_recommendations = generate_recommendations(test_user_item_matrix, item_similarity_df, k=5)

# Generate user-based recommendation lists
user_based_recommendations = generate_user_based_recommendations(user_item_matrix, user_similarity_df, test_users, k=5)

# Merge Recommended Lists
combined_recommendations = combine_recommendations(item_based_recommendations, user_based_recommendations)

# Calculate the accuracy of the combined recommendation results
combined_precision, combined_recall = calculate_accuracy(test_user_item_matrix, combined_recommendations)
combined_precision, combined_recall
```

We can find that the recall has a significant improvement, combine the two recommendation methods can improve the accuracy of the recommendation results. And if we want to sell more products, the accuracy do not need to be too high.
